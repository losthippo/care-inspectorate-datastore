```python
import pandas as pd
import numpy as np
import re
import glob
```

```python
path = r'c:/users/lisa.hendry/Desktop/Github/care-inspectorate-datastore/raw' # use your path
files = glob.glob(path + "/*.csv")

li = []

for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding = "ISO-8859-1", low_memory=False)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True, sort=True)
df['Subtype'] = df['Subtype'].fillna('None')
```

```python
keep = ['CSNumber', 'Combined_Service_', 'CaseNumber_Combined','CareService', 'Subtype', 'ServiceType', \
        'ServiceName', 'Service_Postcode', 'ServiceProvider', 'ServiceStatus', 'DateReg', \
        'Integration_Authority_Name', 'Council_Area_Name', 'Health_Board_Name','Client_group', 'Last_inspection_Date', \
        'Quality_of_Care_and_Support', 'Quality_of_Environment', 'Quality_of_Staffing', 'Quality_of_Mgmt_and_Lship', \
#         'KQ_Support_Wellbeing', 'KQ_Care_and_Support_Planning', 'KQ_Setting', 'KQ_Staff_Team', 'KQ_Leadership', \
#         'Comb_Eval_CareSupport_Wellbeing', 'Comb_Eval_CareSupport_Planning', 'Comb_Eval_Setting_Env', 'Comb_Eval_Staff_StaffTeam', 'Comb_Eval_Leadership_ML', \
        ]
```

```python
counts = df.filter(regex=r'(Complnt|Enforcements|Count any reqs)')
```

```python
data = df[[c for c in df.columns if c in keep]]
```

```python
tps = data[data['ServiceProvider'] == 'Turning Point Scotland'].reset_index(drop=True)
tps.info()
```

```python
tps.to_csv('tps.csv', index=False)
```

```python
stewart = tps[['CSNumber','ServiceName','CareService','Subtype','Last_inspection_Date','Quality_of_Care_and_Support','Quality_of_Environment', 'Quality_of_Mgmt_and_Lship', 'Quality_of_Staffing']]
stewart = stewart.rename({'Quality_of_Care_and_Support':'C&S', 'Quality_of_Environment':'Env', 'Quality_of_Mgmt_and_Lship':'Mgmt', 'Quality_of_Staffing':'Staff', 'Last_inspection_Date':'LastInspection'}, axis=1)
stewart.LastInspection = pd.to_datetime(stewart.LastInspection, dayfirst=True)
stewart = stewart.dropna(thresh=6)

cols = ['C&S','Env','Mgmt','Staff']

stewart[cols] = stewart[cols].fillna(0)

stewart[cols] = stewart[cols].apply(pd.to_numeric, errors='coerce').astype(int)

stewart['Year'] = stewart['LastInspection'].dt.year

stewart = stewart.drop('LastInspection', axis=1).sort_values(['ServiceName','CareService','Year'])

stewart.head()
```

```python
stewartp = stewart.melt(id_vars=['CSNumber','ServiceName','CareService','Subtype','Year'])
stewartp = stewartp.pivot_table(index=['CSNumber', 'ServiceName', 'CareService', 'Subtype'], columns=['Year','variable'], values='value', aggfunc='min').fillna(0).astype(int).reset_index()
stewartp
```

```python
df = pd.concat([data,counts], axis=1)
```

```python
df = df[df['Client_group'] == "Adults"].reset_index(drop=True).drop_duplicates()
df.head()
```

```python
df.Subtype = df.Subtype.fillna('None')
```

```python
servicetypes = ["Support Service", "Care Home Service", "Offender Accommodation Service", "Housing Support Service"] 
```

```python
df = df[df['CareService'].isin(servicetypes)]
```

```python
idvars = [
    'CSNumber', 'CareService', 'CaseNumber_Combined', 'Client_group', \
    'ServiceName', 'ServiceProvider', 'ServiceStatus', 'ServiceType', 'Service_Postcode', 'Subtype', \
    'Combined_Service_', 'Council_Area_Name', 'DateReg', 'Health_Board_Name', 'Integration_Authority_Name', \
    'Last_inspection_Date' 
]
```

```python
dff = df.melt(id_vars=idvars).dropna(subset=['value'])
dff.value = dff.value.replace('Yes', 1)
```

```python
dff.variable.value_counts()
```

```python
def split_year(string):
    m = re.search(r'\d+$', string)
    
    if m is not None:
        return m.group()
        
    else: 
        return np.nan
```

```python
dff['year'] = dff['variable'].apply(split_year)
```

```python
dff['variable'] = dff['variable'].str.replace(r'\d+$', '')
```

```python
dff.year.value_counts()
```
