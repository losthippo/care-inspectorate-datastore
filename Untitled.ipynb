```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('MDSF_data.csv', encoding = "ISO-8859-1")

servicetypes = ["Support Service", "Care Home Service", "Offender Accommodation Service", "Housing Support Service"] 

df = df[df['CareService'].isin(servicetypes)].drop(["Address_line_1","Address_line_2","Address_line_3", "Address_line_4","ManagerName", "Service_Phone_Number", "Eforms_email_address", "Date_Reg"], axis=1)
```

```python
id_vars = ["CSNumber", 
"Combined_Service_",
"CaseNumber_Combined",
"CareService",
"Subtype",
"ServiceType",
"ServiceName",
"Service_town",
"Service_Postcode",
"SP_number",
"ServiceProvider",
"Provided_by_Local_Authority",
"ServiceStatus",
"DateReg"]

df = pd.melt(frame=df, id_vars=id_vars, var_name='Measure', value_name='Value')
df.head()
```
